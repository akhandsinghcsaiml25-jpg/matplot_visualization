```python
students = {
    1: {"name": "Aman", "study_hours": 5, "attendance": 85, "marks": 82},
    2: {"name": "Riya", "study_hours": 3, "attendance": 70, "marks": 65},
    3: {"name": "Rahul", "study_hours": 2, "attendance": 80, "marks": 35},
    4: {"name": "Sneha", "study_hours": 1, "attendance": 60, "marks": 18},
    5: {"name": "Vikram", "study_hours": 4, "attendance": 90, "marks": 76}
}

for student_id, details in students.items():
    print("\nStudent:", details["name"])

    marks = details["marks"]
    attendance = details["attendance"]
    study_hours = details["study_hours"]

    if marks > 75:
        print("Performance: Strong")
    elif marks < 20:
        print("Performance: Fail - Repeat Class")
    elif marks <= 40:
        print("Performance: Needs Improvement")
    else:
        print("Performance: Average")

    if attendance < 75:
        print("Attendance Status: Low Attendance")


    if study_hours < 4:
        print("Warning: Needs More Study Time")

    print("Marks:", marks)
    print("Attendance:", attendance, "%")
    print("Study Hours:", study_hours)
```

```python
import pandas as pd
import numpy as np

# Read CSV file
df = pd.read_csv("session4_dummy_students_data.csv.xls")

# Convert columns to NumPy arrays
math_marks = df["math"].to_numpy()
science_marks = df["science"].to_numpy()
english_marks = df["english"].to_numpy()

# Calculate total marks
total_marks = math_marks + science_marks + english_marks

# Display results
for i in range(len(df)):
    print(
        df["name"][i],
        math_marks[i],
        science_marks[i],
        english_marks[i],
        total_marks[i]
    )
```

```python
df.head()
```

```python


# Create average column
df["average"] = np.mean(
    df[["math", "science", "english"]],
    axis=1
)

# Create result column
df["result"] = np.where(
    df["average"] >= 40,
    "Pass",
    "Fail"
)

# Print updated dataframe
print("Student Data:")
print(df)

# Class average
class_avg = np.mean(df["average"])
print("\nClass Average =", class_avg)

# Highest and Lowest marks subject wise
subjects = ["math", "science", "english"]

for sub in subjects:
    print(f"\n{sub.upper()}")
    print("Highest =", np.max(df[sub]))
    print("Lowest =", np.min(df[sub]))

# Top student(s)
max_avg = np.max(df["average"])
top_students = df[df["average"] == max_avg]

print("\nTop Student(s):")
print(top_students[["name", "average"]])

# Attendance below 75
print("\nStudents with attendance below 75:")
print(df[df["attendance_percent"] < 75][["name", "attendance_percent"]])

# Assignments less than 6
print("\nStudents who submitted fewer than 6 assignments:")
print(df[df["assignments_submitted"] < 6][
    ["name", "assignments_submitted"]
])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)


print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

```python
print("Minimum Year:", df["year"].min())
print("Maximum Year:", df["year"].max())
print("range", df["year"].min(),"-",df["year"].max())
num_countries = df["country"].nunique()
print("Number of Countries:", num_countries)
num_continents = df["continent"].nunique()
print("Number of Continents:", num_continents)
print(df["continent"].unique())
# Total records
print(df.shape[0])

# Total columns
print(df.shape[1])

# Average life expectancy
print(df["lifeExp"].mean())

# Highest GDP per capita
print(df["gdpPercap"].max())

# Country count by continent
print(df.groupby("continent")["country"].nunique())
```

```python

import matplotlib.pyplot as plt




life_exp = df.groupby("continent")["lifeExp"].mean()

print(life_exp)

plt.figure(figsize=(8,5))
plt.bar(life_exp.index, life_exp.values)

plt.title("Average Life Expectancy by Continent")
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")

plt.show()
```

```python

import matplotlib.pyplot as plt


life_exp = df.groupby("continent")["lifeExp"].sum()

plt.bar(life_exp.index, life_exp.values)
plt.xlabel("Continent")
plt.ylabel("Total Life Expectancy")
plt.title("Total Life Expectancy by Continent")
plt.show()
```

```python

# Select countries
countries = ["India", "China", "United States"]

# Filter data
filtered_df = df[df["country"].isin(countries)]

# Plot
plt.figure(figsize=(10, 8))

for country in countries:
    country_data = filtered_df[filtered_df["country"] == country]
    plt.plot(country_data["year"],
             country_data["lifeExp"],
             marker='o',
             label=country)

plt.title("Life Expectancy Over Time")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend()
plt.grid(True)

plt.show()
```
